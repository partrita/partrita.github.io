# Dropping Columns in a DataFrame


```python
df = pd.read_csv('https://raw.githubusercontent.com/realpython/python-data-cleaning/master/Datasets/BL-Flickr-Images-Book.csv')
df.tail()
```

We can drop these columns in the following way:

```python
to_drop = ['Edition Statement',
            'Corporate Author',
            'Corporate Contributors',
            'Former owner',
            'Engraver',
            'Contributors',
            'Issuance type',
            'Shelfmarks']

df.drop(columns=to_drop, inplace=True)
```

When we inspect the DataFrame again, we’ll see that the unwanted columns have been removed:

```python
df.tail()
```
